    Q3 : Selectioner un sous-ensemble de DF, dont les valeurs de la colonne 'Coeff' sont dans compris entre 2 et 8 (indication : methode .between()).

```python
import pandas as pd
df = pd.DataFrame([['Math', 'en ligne', 2, False],['Data Processing','presentiel', 9, True],['Python', 'en ligne', 6, True],['SGBDR', 'en ligne', 5, False],['BD Spatiale', 'presentiel',4, False],['Rech Biblio', 'mixte', 8, True],['IA', 'presentiel', 4, True]])
df
```

```python
df.columns = ['Module', 'Modal_Cours', 'Coeff', 'Valider']
df
```

```python
df1  = df['Modal_Cours'].isin(['presentiel','en ligne'])
df[df1]
```

```python
df2= df["Coeff"].between(2, 8, inclusive = True)
df[df2]
```
